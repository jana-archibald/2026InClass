## `prop.table` makes tables with proportions.  There is an option that chooses different proportion tables

### The `1` makes a proportion by row table

```r
prop.table(table(titanicdata$Sex,titanicdata$Class),1)
```

### The `2` makes a proportion by column table

```r
prop.table(table(titanicdata$Sex,titanicdata$Class),2)
```

## Making bar plots from contingency tables

```r
barplot(table(titanicdata$Sex,titanicdata$Class),legend=T)
```

```r
barplot(table(titanicdata$Sex,titanicdata$Class),legend=T,beside=T)
```

### Making plots that show proportions.

```r
barplot(prop.table(table(titanicdata$Sex,titanicdata$Class),2),legend=T)
```

### Using R to make a mosaic plot

There are many options for making mosaic plots.

I read the one below as `Survival` split by `Sex`.

First I am using an approach that is similar to what we use with `lm` for linear regression.

```r
mosaicplot(Sex~Survived,data=titanicdata,col=T)
```

The next block uses an approach similar to what is shown in lab 5.  First making a table, and then asking R to plot the table.

```r
mosaicplot(table(titanicdata$Sex,titanicdata$Survived),col=T)
```

Now we spent some time exploring what happens when you change the order in the table.  (You can also try this with the first approach!) I always encourage trying something and looking carefully at what it produces.

```r
table(titanicdata$Survived,titanicdata$Sex)
```

```r
mosaicplot(table(titanicdata$Survived,titanicdata$Sex),col=T)
```

```r
mosaicplot(t(table(titanicdata$Survived,titanicdata$Sex)),col=T,main="Make a better title")
```

## Comparing the multiple plot options.

```r
par(mfrow=c(2,2))
barplot(table(titanicdata$Survived,titanicdata$Sex),main="Barplot of Survival and Sex",legend=T,ylab="Frequency",beside=T)
barplot(table(titanicdata$Survived,titanicdata$Sex),main="Barplot of Survival and Sex",legend=T,ylab="Frequency")
barplot(prop.table(table(titanicdata$Survived,titanicdata$Sex),2),main="Barplot of Survival and Sex",legend=T,ylab="Frequency")
mosaicplot(Sex~Survived,data=titanicdata,col=T)
```
